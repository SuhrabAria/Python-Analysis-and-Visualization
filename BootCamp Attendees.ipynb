```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "C:/Users/Suhrab Aria/Documents/Jupyter notebook/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
project = pd.read_csv(file)
project.head()
```

```python
# Inspect all columns
project.describe()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
specific_columns = project.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
specific_columns.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
replaced_values = specific_columns.replace([0.0, 1], ["No", "Yes"])
replaced_values.head()
```

```python
# Calculate total number of respondents in survey
replaced_values['AttendedBootcamp'].value_counts()
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended = replaced_values[replaced_values['AttendedBootcamp'] == "Yes"]
attended.head()
```

```python
# Calculate average age of attendees
average_age = replaced_values["Age"].mean()
round_age = round(average_age)
# Calculate how many people attended a bootcamp
people_attended = attended['AttendedBootcamp'].count()
# Calculate how many attendees hold degrees
attendees_degrees = replaced_values['SchoolDegree'].value_counts().sum()
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender = replaced_values['Gender'].value_counts().sum()
# Calculate percentage of respondents who attended a bootcamp
replaced_values['AttendedBootcamp'].value_counts()
percentage = replaced_values['AttendedBootcamp'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
yes_responds = percentage.loc["Yes"]
# Calculate average post-bootcamp salary
average_salary = replaced_values['BootcampPostSalary'].mean()
round_average = round(average_salary)

Bootcamp_dataframe = pd.DataFrame({'Average Age of Attendees': [round_age],
                                  'Number of People Attended Bootcamp': [people_attended],
                                  'Number of Attendees hold degrees': [attendees_degrees],
                                  'Number of Attendees with Gender Identified': [total_gender],
                                  'Percentage of People attended bootcamp': [yes_responds],
                                  'Average Post-Bootcamp Salary': [round_average]})
Bootcamp_dataframe
```

```python
# Calculate percentage of respondents belonging to each gender
replaced_values['Gender'].value_counts()
gender_percentage = replaced_values['Gender'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'

above_dataframe = pd.DataFrame({'Gender Percentage': gender_percentage})
above_dataframe
```

```python
# Calculate percentage of attendees with a school degree
replaced_values['SchoolDegree'].value_counts()
degree_percentage = replaced_values['SchoolDegree'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'

above_dataframe2 = pd.DataFrame({'Attendees with School Degree': degree_percentage})
above_dataframe2
```
